```python
import numpy as np
import pandas as pd
```

```python
'''
target value:
0 - stay
1 - walk
2 - run
'''

def preprocess_df(df, target : int, final_df = None, flag = False, idx_0 = None, idx_1 = None):
  df = df.rename(columns={'v1' : 'gyro_x', 
                          'v2' : 'gyro_y',
                          'v3' : 'gyro_z',
                          'v4' : 'acceleration_x',
                          'v5' : 'acceleration_y',
                          'v6' : 'acceleration_z'})
  
  df = df.drop(columns = ['pocket_id', 'v7'])
  df[['target']] = target
  len_df = len(df)

  if idx_0:
    df = df.iloc[idx_0:]

  if idx_0 and idx_1:
    df = df.iloc[:idx_1 - len_df]

  if not idx_0 and idx_1:
    df = df.iloc[:idx_1]
    
  if idx_0 or idx_1:
    df = df.reset_index(drop = True)

  if flag:
    final_df = pd.concat([final_df, df], ignore_index = True)
    return df, final_df

  return df
```

```python
stay_df_0 = pd.read_csv('./stay_0.csv')
stay_df_0 = preprocess_df(stay_df_0, target = 0)
final_df = stay_df_0
stay_df_0
```
